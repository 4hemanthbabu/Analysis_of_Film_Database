### Analysing Movies on the Basis of Year

```python
year = pa.unique(File['Year'])
year.sort()
year = nu.array(year)
year
```

```python
d = File['Year'].value_counts()
print(d)
y = []
for item in year:
    val = d[item]
    y.append(val)
print(y)
```

```python
fig_dims = (15, 8)
fig, ax = plt.subplots(figsize=fig_dims)
line = sea.lineplot(x = year, y = y)
line_plot = line.set(xlabel = 'Year', ylabel = 'Number of Movies', title = 'Year wise Movie Release')
```

Year 1991, 129 movies were released and years 1920, 1923, 1925,1996 and 1997 only one movie were released according to the dataset

## Analysing Actor on the basis of Type of movies

```python
Subjects = pa.unique(File['Subject'])
Subjects
```

```python
D = {}
for sub in Subjects:
    D[sub] = []
    for yr in year:
        count = File.groupby(['Year', 'Subject']).size()
        if yr in count and sub in count[yr]:
            val = count[yr][sub]
        else:
            val = 0
        D[sub].append(val)
print(D)
```

```python
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ax.set_title( "Categorical Movie Behaviour" , size = 24 )
for sub in Subjects:
    sea.lineplot(x = year, y = D[sub],
                 label = sub)
plt.legend(bbox_to_anchor=(1.05, 1), loc = 2, borderaxespad = 0, title = 'TYPE') 
plt.show()
                 

```

## Actor wise Analysis

```python
Actor = pa.unique(File['Actor'])
y= []
Actor_count = File.groupby('Actor').size().sort_values(ascending=False)[:10]
x = Actor_count.keys()
for item in x:
    val = Actor_count[item]
    y.append(val)
print(y)
```

### 10 Actors with Most Number of Movies

```python
fig_dims = (15, 3)
fig, ax = plt.subplots(figsize=fig_dims)
Actor_plot = sea.barplot(x = x, y = y)
Actor_plot.set(xlabel = 'Actors', ylabel = 'Number of Movies')
Actor_plot.set_title('10 Actors with Most Number of Movies',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.show()
```

## Analysis of Movies with Actress Column 

```python
Actress = pa.unique(File['Actress'])
Actress
len(Actress)
```

### 10 Actress with Most Number of Films

```python
y = []
Actress_count = File.groupby('Actress').size().sort_values(ascending = False)
x = Actress_count.keys()
for ats in x:
    val = Actress_count[ats]
    y.append(val)
print(y)
```

```python
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
Actress_histogram = sea.distplot([y])
Actress_histogram.set(xlabel = 'Number of Movies', ylabel = 'Frequency')
Actress_histogram.set_title('Actress Movie Distribution', fontdict = {'fontsize' : 24, 'fontweight': 'bold'})
plt.show()
```

## Analysis of Director and Films

```python
Directors = pa.unique(File['Director'])
Directors
len(Directors)
```

```python
y = []
#Dir_pop = File.groupby(['Director','Popularity'])
Dir_pop = File[['Director','Popularity']]
# print(Dir_pop)
Dir_pop_mean = Dir_pop.groupby('Director').describe().reset_index()
#print(Dir_pop_mean)
#print(Dir_pop_mean.columns)
t = Dir_pop_mean.sort_values(by = ('Popularity', 'mean'), ascending = False)
#print(t)
l = t.loc[t[('Popularity', 'count')] >= 5][:10]
print(l.columns)


#Dir_pop_std = Dir_pop.groupby('Director').std()
#Dir_pop_std
```

```python
fig_dims = (20, 12)
fig, ax = plt.subplots(figsize=fig_dims)
Dir_pop =  sea.barplot(x = x, y = y,)
Dir_pop.errorbar(range(len(x)), y, yerr = z, fmt = 'o', color = 'k')
Dir_pop.set(xlabel = 'Directors', ylabel = 'Popularity')
Dir_pop.set_title('10 Directors having Most Popularity',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.show()
```
